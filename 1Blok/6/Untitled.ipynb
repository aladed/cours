```python
df10 = pd.read_sql("""
    SELECT u.city, COUNT(f.action)
    FROM "user" u
    JOIN feed_action f ON f.user_id = u.id 
    JOIN post p ON p.id = f.post_id
    WHERE topic = 'covid' AND age = 36 AND DATE(f.time) = '2021-12-01' AND action ='view'
    GROUP BY u.city 
    ORDER BY COUNT(f.action)
    """,
    con=conn_url
)
df10
```

```python
SELECT 
    post_id,
    COUNT(DISTINCT user_id) AS unique_likes,
    MAX(time) AS last_like_time
FROM feed_action
WHERE  action = 'like'
GROUP BY post_id
ORDER BY unique_likes DESC
LIMIT 5
```
